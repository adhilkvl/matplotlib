## axis labels, legend, grid

### axis label

```python
day=[1,2,3,4,5,6,7]
max_temp=[50,51,59,48,47,49,58]
min_temp=[43,42,40,44,33,35,37]
avg_temp=[45,48,48,46,40,42,41]
```

```python
plt.title("week one",loc="center")
plt.xlabel("days")
plt.ylabel("temperature")
plt.plot(day,max_temp)
plt.plot(day,min_temp)
plt.plot(day,avg_temp)
plt.show()
```

### legend

```python
plt.title("week one",loc="center")
plt.xlabel("days")
plt.ylabel("temperature")
plt.plot(day,max_temp,label="max")
plt.plot(day,min_temp,label="min")
plt.plot(day,avg_temp,label="avg")
plt.legend(loc="lower left",shadow=True)
plt.show()
```

### grid

```python
plt.title("week one",loc="center")
plt.xlabel("days")
plt.ylabel("temperature")
plt.plot(day,max_temp,label="max")
plt.plot(day,min_temp,label="min")
plt.plot(day,avg_temp,label="avg")
plt.legend(loc="lower left",shadow=True)
plt.grid()
plt.show()
```

## Bar chart

```python
company=["google","amazon","flipcart","spotify"]
revenue=[101,201,253,154]
profit=[50,25,155,88]
```

```python
ypos=np.arange(len(company))
```

```python
plt.ylabel("revenue(bill)")
plt.title("Us stocks")
plt.xticks(ypos,company)
plt.bar(ypos,revenue,width=0.4,label="revenue")
plt.bar(ypos,profit,width=0.4,label="profit")
plt.legend(shadow=True)
plt.show()
```

## vertical bar chart

```python
plt.ylabel("revenue(bill)")
plt.title("Us stocks")
plt.xticks(ypos,company)
plt.bar(ypos-0.2,revenue,width=0.4,label="revenue")
plt.bar(ypos+0.2,profit,width=0.4,label="profit")
plt.legend(shadow=True)
plt.show()
```

## horizontal bar chart

```python
plt.xlabel("revenue(bill)")
plt.title("Us stocks")
plt.yticks(ypos,company)
plt.barh(ypos-0.2,revenue,height=0.4,label="revenue")
plt.barh(ypos+0.2,profit,height=0.4,label="profit")
plt.legend(shadow=True)
plt.show()
```

## subplot

```python
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])
#plot 1:
plt.subplot(1, 2, 1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)

plt.show()
```

```python
plt.subplot(1, 2, 1)
plt.xlabel("revenue(bill)")
plt.title("Us stocks")
plt.yticks(ypos,company)
plt.barh(ypos-0.2,revenue,height=0.4,label="revenue")
plt.barh(ypos+0.2,profit,height=0.4,label="profit")
plt.legend(shadow=True)

plt.subplot(1, 2, 2)
plt.ylabel("revenue(bill)")
plt.title("Us stocks")
plt.xticks(ypos,company)
plt.bar(ypos-0.2,revenue,width=0.4,label="revenue")
plt.bar(ypos+0.2,profit,width=0.4,label="profit")
plt.legend(shadow=True)


plt.suptitle("Subplot")
plt.tight_layout()
plt.show()
```


## Histograms

```python
blood_sugar_men=[113,85,90,150,149,88,93,115,135,80,77,82,129]
blood_sugar_women=[110,95,80,150,94,88,93,115,145,80,77,82,139]
```

```python
plt.title("Blood suger analysis")
plt.xlabel("suger level")
plt.ylabel("total patients")
plt.hist([blood_sugar_men,blood_sugar_women],bins=[80,100,125,150],rwidth=.95,color=['g','b'],label=["men","women"],width=9)
plt.legend()
plt.show()
```

## horizontal hist

```python
plt.title("Blood suger analysis")
plt.xlabel("total patients")
plt.ylabel("suger leval")
plt.hist([blood_sugar_men,blood_sugar_women],bins=[80,100,125,150],rwidth=.95,color=['g','b'],label=["men","women"]
         ,orientation='horizontal',height=9)
plt.legend()
plt.show()
```

## pie chart

```python
exp_value=[1400,750,1200,850,600]
exp_label=["home rent","bill","car","phone bill","other utilities"]
```

```python
plt.pie(exp_value,labels=exp_label,radius=1.2,autopct='%0.1f%%')
plt.show()
```

## pie chart 2

```python
plt.pie(exp_value,radius=1.2,autopct='%0.1f%%',shadow=True,explode=[0,0,0.1,0,0])
plt.legend(exp_label,bbox_to_anchor=(1,1))
plt.show()
```
